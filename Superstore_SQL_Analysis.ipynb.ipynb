```python
# Step 1: Load CSV into SQLite Database
df = pd.read_csv('Sample - Superstore.csv', encoding='latin1')
conn = sqlite3.connect(':memory:')
df.to_sql('superstore', conn, if_exists='replace', index=False)
print('Data loaded successfully!')
print(f'Total rows: {len(df)}')
print(f'Total columns: {len(df.columns)}')
```

## Step 2: Explore Table â Schema & Sample Data

```python
# Column names and data types
print('=== Column Names & Data Types ===')
print(df.dtypes)
```

```python
# Sample first 5 rows
pd.read_sql_query('SELECT * FROM superstore LIMIT 5', conn)
```

## Step 3: WHERE Filters(region, category, date, sales)

```python
# Filter by Region = West
pd.read_sql_query('''
SELECT "Order ID", "Customer Name", Region, Sales
FROM superstore
WHERE Region = 'West'
LIMIT 10
''', conn)
```

```python
# Filter by Category = Technology
pd.read_sql_query('''
SELECT "Order ID", "Customer Name", Category, Sales
FROM superstore
WHERE Category = 'Technology'
LIMIT 10
''', conn)
```

```python
# Filter by Year = 2017
pd.read_sql_query('''
SELECT "Order ID", "Order Date", Sales
FROM superstore
WHERE "Order Date" LIKE '%2017%'
LIMIT 10
''', conn)
```

```python
# High Value Orders: Sales > 1000
pd.read_sql_query('''
SELECT "Order ID", "Customer Name", Sales, Profit
FROM superstore
WHERE Sales > 1000
ORDER BY Sales DESC
LIMIT 10
''', conn)
```

## Step 4: GROUP BY Aggregations(sales, quantity, averages)

```python
# Total Sales by Region
pd.read_sql_query('''
SELECT Region,
       ROUND(SUM(Sales), 2) AS Total_Sales,
       ROUND(AVG(Sales), 2) AS Avg_Sales,
       SUM(Quantity)        AS Total_Quantity
FROM superstore
GROUP BY Region
ORDER BY Total_Sales DESC
''', conn)
```

```python
# Sales & Profit by Category
pd.read_sql_query('''
SELECT Category,
       ROUND(SUM(Sales), 2)  AS Total_Sales,
       ROUND(SUM(Profit), 2) AS Total_Profit,
       COUNT(*)              AS Order_Count
FROM superstore
GROUP BY Category
ORDER BY Total_Sales DESC
''', conn)
```

```python
# Sub-Category with Profit %
pd.read_sql_query('''
SELECT "Sub-Category",
       ROUND(SUM(Sales), 2)                      AS Total_Sales,
       ROUND(SUM(Profit), 2)                     AS Total_Profit,
       ROUND(SUM(Profit)*100.0/SUM(Sales), 1)    AS Profit_Pct
FROM superstore
GROUP BY "Sub-Category"
ORDER BY Total_Sales DESC
''', conn)
```

##### Step 5: Sort & Limit â Top Products & categories

```python
# Top 10 Products by Sales
pd.read_sql_query('''
SELECT "Product Name",
       ROUND(SUM(Sales), 2) AS Total_Sales,
       SUM(Quantity)        AS Total_Quantity
FROM superstore
GROUP BY "Product Name"
ORDER BY Total_Sales DESC
LIMIT 10
''', conn)
```

```python
# Top 5 Categories by Sales
pd.read_sql_query('''
SELECT Category,
       ROUND(SUM(Sales), 2) AS Total_Sales,
       ROUND(SUM(Profit), 2) AS Total_Profit,
       COUNT(*) AS Order_Count
FROM superstore
GROUP BY Category
ORDER BY Total_Sales DESC
LIMIT 5
''', conn)
```

## Step 6:solve use case (monthly trends , top customer, duplicates)

```python
# Monthly Sales Trend
pd.read_sql_query('''
SELECT substr("Order Date", -4) AS Year,
       CASE substr("Order Date", 1, instr("Order Date","/")-1)
           WHEN '1' THEN '01-Jan' WHEN '2' THEN '02-Feb' WHEN '3' THEN '03-Mar'
           WHEN '4' THEN '04-Apr' WHEN '5' THEN '05-May' WHEN '6' THEN '06-Jun'
           WHEN '7' THEN '07-Jul' WHEN '8' THEN '08-Aug' WHEN '9' THEN '09-Sep'
           WHEN '10' THEN '10-Oct' WHEN '11' THEN '11-Nov' WHEN '12' THEN '12-Dec'
       END AS Month,
       ROUND(SUM(Sales), 2) AS Monthly_Sales
FROM superstore
GROUP BY Year, Month
ORDER BY Year, Month
''', conn)
```

```python
# Top 10 Customers by Sales
pd.read_sql_query('''
SELECT "Customer Name", Segment, Region,
       ROUND(SUM(Sales), 2)  AS Total_Sales,
       ROUND(SUM(Profit), 2) AS Total_Profit
FROM superstore
GROUP BY "Customer Name"
ORDER BY Total_Sales DESC
LIMIT 10
''', conn)
```

```python
# Duplicate Order IDs
pd.read_sql_query('''
SELECT "Order ID", COUNT(*) AS Row_Count
FROM superstore
GROUP BY "Order ID"
HAVING COUNT(*) > 1
ORDER BY Row_Count DESC
LIMIT 10
''', conn)
```

```python
# Loss-making Sub-Categories
pd.read_sql_query('''
SELECT "Sub-Category",
       ROUND(SUM(Sales), 2)  AS Total_Sales,
       ROUND(SUM(Profit), 2) AS Total_Profit
FROM superstore
GROUP BY "Sub-Category"
HAVING SUM(Profit) < 0
ORDER BY Total_Profit
''', conn)
```

## Step 7: Validate Results

```python
# Row count and null check
pd.read_sql_query('''
SELECT COUNT(*)       AS Total_Rows,
       COUNT(Sales)   AS Sales_NonNull,
       COUNT(Profit)  AS Profit_NonNull,
       COUNT(Region)  AS Region_NonNull
FROM superstore
''', conn)
```

```python
# Distinct values check
pd.read_sql_query('''
SELECT COUNT(DISTINCT "Order ID")      AS Unique_Orders,
       COUNT(DISTINCT "Customer Name") AS Unique_Customers,
       COUNT(DISTINCT "Product Name")  AS Unique_Products,
       COUNT(DISTINCT Region)          AS Regions,
       COUNT(DISTINCT Category)        AS Categories
FROM superstore
''', conn)
```
