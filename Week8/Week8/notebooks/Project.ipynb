```python
import pandas as pd
from faker import Faker
import random

fake = Faker()

print(fake.name())
print(fake.email())
```

```python
customers = []

for i in range(1000):
    customers.append({
        "customer_id": f"C{i+1:03}",
        "customer_name": fake.name(),
        "email": fake.email(),
        "registration_date": fake.date_between(start_date="-2y", end_date="today"),
        "customer_type": random.choice(["REGULAR", "PREMIUM", "VIP"])
    })

customers
```

```python
customers_df = pd.DataFrame(customers)

customers_df
```

```python
customers_df.to_csv("customers.csv", index=False)

print("customers.csv created successfully!")
```

```python
product_names = [
    "Laptop", "Mobile", "Headphones", "Keyboard", "Mouse",
    "T-Shirt", "Jeans", "Shoes", "Jacket", "Watch",
    "Book", "Notebook", "Chair", "Table", "Bottle",
    "Bag", "Camera", "Speaker", "Monitor", "Printer"
]

categories = {
    "Electronics": ["Laptop", "Mobile", "Headphones", "Keyboard", "Mouse", "Camera", "Speaker", "Monitor", "Printer"],
    "Clothing": ["T-Shirt", "Jeans", "Shoes", "Jacket"],
    "Books": ["Book", "Notebook"],
    "Home": ["Chair", "Table", "Bottle", "Bag"]
}
```

```python
products = []

for i in range(500):

    category = random.choice(list(categories.keys()))
    product = random.choice(categories[category])

    products.append({
        "product_id": f"P{i+1:03}",
        "product_name": product,
        "category": category,
        "subcategory": product,
        "cost_price": random.randint(100,50000)
    })

products_df = pd.DataFrame(products)

products_df.head()
```

```python
products_df.to_csv("products.csv", index=False)

print("products.csv created successfully!")
```

```python
statuses = ["PLACED", "SHIPPED", "DELIVERED", "CANCELLED", "RETURNED"]

regions = ["North", "South", "East", "West"]
```

```python
orders = []

for i in range(10000):

    orders.append({

        "order_id": f"O{i+1:05}",

        "customer_id": f"C{random.randint(1,1000):03}",

        "order_date": fake.date_time_between(start_date="-2y", end_date="now"),

        "status": random.choice(statuses),

        "region_code": random.choice(regions)

    })

orders_df = pd.DataFrame(orders)

orders_df.head()
```

```python
order_items = []

for i in range(30000):

    order_items.append({

        "item_id": f"I{i+1:05}",

        "order_id": f"O{random.randint(1,10000):05}",

        "product_id": f"P{random.randint(1,500):03}",

        "quantity": random.randint(1,5),

        "unit_price": random.randint(100,50000),

        "discount_percent": random.randint(0,50)

    })

order_items_df = pd.DataFrame(order_items)

order_items_df.head()
```
