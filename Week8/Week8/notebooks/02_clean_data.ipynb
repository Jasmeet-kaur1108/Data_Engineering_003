```python
# Import required libraries
import pandas as pd
import numpy as np
```

```python
# Load raw CSV files
customers = pd.read_csv("../data/raw/customers.csv")
products = pd.read_csv("../data/raw/products.csv")
orders = pd.read_csv("../data/raw/orders.csv")
order_items = pd.read_csv("../data/raw/order_items.csv")

print("Customers :", customers.shape)
print("Products :", products.shape)
print("Orders :", orders.shape)
print("Order Items :", order_items.shape)
```

```python

print(customers.isnull().sum())

print(products.isnull().sum())

print(orders.isnull().sum())

print(order_items.isnull().sum())
```

```python
# 5% customer_id to null conversion

orders.loc[orders.sample(frac=0.05).index, "customer_id"] = None
```

```python
# 2% invalid emails

customers.loc[customers.sample(frac=0.02).index, "email"] = "invalid_email"
```

```python
# Extra spaces in products names

products.loc[products.sample(frac=0.05).index, "product_name"] = "  laptop  "
```

```python
# 3% negative quantity

order_items.loc[order_items.sample(frac=0.03).index, "quantity"] *= -1
```

```python
print(orders.isnull().sum())

print(customers.head())

print(products.head())

print(order_items.head())
```

```python
customers["email"] = customers["email"].replace("invalid_email", "unknown@example.com")
```

```python
products["product_name"] = products["product_name"].str.strip().str.title()
```

```python
orders["customer_id"] = orders["customer_id"].fillna("Unknown")
```

```python
order_items["quantity"] = order_items["quantity"].abs()
```

```python
# Remove duplicate records
customers = customers.drop_duplicates()
products = products.drop_duplicates()
orders = orders.drop_duplicates()
order_items = order_items.drop_duplicates()
```

```python
# Save cleaned datasets
customers.to_csv("customers_clean.csv", index=False)
products.to_csv("products_clean.csv", index=False)
orders.to_csv("orders_clean.csv", index=False)
order_items.to_csv("order_items_clean.csv", index=False)

print("All cleaned files saved successfully!")
```
